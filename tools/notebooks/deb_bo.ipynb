```python
#!/usr/bin/env python
# coding: utf-8
import sys
import numpy as np
from ase.io.trajectory import TrajectoryWriter,Trajectory
from irff.AtomDance import AtomDance
from ase.calculators.singlepoint import SinglePointCalculator
from ase.io import read,write
from ase import units,Atoms
from ase.visualize import view
from irff.irff_np import IRFF_NP
# from irff.tools import deb_energy
import matplotlib.pyplot as plt
from irff.plot.deb_bde import deb_bp,deb_bo # ,deb_vdw
# get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
# images = Trajectory(args.traj)
images = []
r = 1.0 
rmax=2.5
while r<rmax:
      r += 0.1
      atoms = Atoms('CC',
                    positions=[(0, 0, 0), (r, 0, 0)],
                    cell=[10.0, 10.0, 10.0],
                    pbc=[1, 1, 1])
      images.append(atoms)
deb_bo(images,i=0,j=1,show=True,more=True,x_distance=True,print_=False,nn=True)
```
